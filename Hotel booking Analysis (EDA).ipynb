# DATA ANALYSIS AND VISUALIZATIONS

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5,4))
plt.title('Reservation status count')
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

Findings:

##### RESERVATION STATUS IN TWO DIFFERENT TYPES OF HOTEL 

```python
plt.figure(figsize=(8,4))
axl=sns.countplot(x='hotel',hue='is_canceled',data=df)
legend_labels,_=axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels',size=20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

 Findings:
  City hotels has more bookings and cancellations than resort type, reason   may be resort hotels are more expensive
 

##### Find the ratio of cancelled and not cancelled reservation bookings in resort hotel

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

##### Find the ratio of cancelled and not cancelled reservation booking in city hotel

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel= resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City and Resort Hotel ',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label ='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

Findings: 
The line graph above shows that,city hotels is in mid(ADR),some spikes may be due to weekend_holidays. The average daily rate for a city hotel is less than that of a resort hotel

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
axl=sns.countplot(x='month',hue='is_canceled',data=df,palette='bright')
legend_labels,_=axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title('Resevation status per month',size=20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not_canceled','canceled'])
plt.show()
```

We have developed the grouped bar graph to analyze the months with the highest and lowest reservation levels according to reservation status.
The number of confirmed reservations are highest in the month of August.
lowest cancellation  in the month of August.
January is the month with the most cancelled reservations
 

```python
plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=30)
df_filtered=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()
sns.barplot(x='month',y='adr',data=df_filtered)
plt.show() 
```

This bar graph demonstrates that cancellations are most common when prices are greatest and are least common when they are lowest.Therefore,the cost of the accomodation is soley responsible for the cancellation.

##### 2. From where most of the customers resevations are coming?
Is it coming from  offline agents,online,Direct or groups?

```python
df['market_segment'].value_counts()
```

```python
df['market_segment'].value_counts(normalize=True)
```

```python
cancelled_data=df[df['is_canceled']==1]
cancelled_data['market_segment'].value_counts(normalize=True)
```

Findings:
Almost 47% customers coming through online agents cancel thier booking.

```python
cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace =True)

not_cancelled_data=df[df['is_canceled']==0]
not_cancelled_df_adr=not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace = True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend()
```

```python
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016')& (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016')& (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate',fontsize=30)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend(fontsize=20)
plt.show()
```

 We see that the cancelled line(orange) is more than the not cancelled one.
  It proves that the prices have effect on cancellation.
  Therefore,ADR is the most influencing factor on the cancellation rate.
